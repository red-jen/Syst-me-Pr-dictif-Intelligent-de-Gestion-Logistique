# Ingestion et Préparation des Données

## Création d’une session Spark et chargement du dataset logistique

```python
import os
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("gestionlogistique").getOrCreate()

df = spark.read.csv("/src/data/DataCoSupplyChainDataset.csv", header=True, inferSchema=True, encoding="ISO-8859-1")
df.show(100)
df.printSchema()
```

## Sélection des colonnes pertinentes du jeu de données logistique : V1

```python
important_cols = [
    "Days for shipping (real)",
    "Days for shipment (scheduled)",
    "Benefit per order",
    "Sales per customer",
    "Late_delivery_risk",
    "Department Id",
    "Latitude",
    "Longitude",
    "Order Item Quantity",
    "Order Item Product Price",
    "Order Item Discount",
    "Order Item Total",
    "Order Profit Per Order",
    "Type",
    "Delivery Status",
    "Category Id",
    "Customer City",
    "Customer State",
    "Order City",
    "Order State",
    "Order Region",
    "Shipping Mode",
    "Order Country"
]
df_clean = df.select(*important_cols)
df_clean.show(5)
```

## Suppression des doublons dans le DataFrame

```python
df = df_clean.dropDuplicates()
```

## Comptage des valeurs nulles par colonne

```python
for i in df.columns:
    null_count = df_clean.filter(df_clean[i].isNull()).count()
    print(f"{i}: {null_count} valeurs nulles")
```

## Récupération des coordonnées géographiques des villes

```python
import requests
import time

city_coords = {}
count = 0

all_locations = df.select("Order City","Customer State").distinct().collect()

for row in all_locations:
    city = row["Customer City"]
    state = row["Customer State"]

    try:
        response = requests.get(
            f"https://api.geoapify.com/v1/geocode/search?text={city},{state}&apiKey=b36185c337a14a57ba7ce5a36e5bfffc",
            timeout=10
        )
        data_city = response.json()
        features = data_city.get("features", [])
        if features:
            city_data = features[0]
            latitude = city_data["properties"].get("lat", 0)
            longitude = city_data["properties"].get("lon", 0)
        else:
            latitude, longitude = 0, 0

    except Exception as e:
        print(f"Error for {city}, {state}: {e}")
        latitude, longitude = 0, 0

    city_coords[city] = (latitude, longitude)
    count += 1
    print(count, latitude, longitude)

    time.sleep(0.1)
```

## Résoudre les villes non géocodées manuellement

```python

for i,j in city_coords.items():
    if j[0]==0 and j[1]==0:
        print(i)
```

```python
name_corrections = {
    "Aew?l-li": "Aewŏl-li",
    "Klaip?da": "Klaipėda",
    "Kahramanmara?": "Kahramanmaraş",
    "Bra?ov": "Brașov",
    "Gy?r": "Győr"
}

coordinates = {
    "Aewŏl-li": (33.395, 126.288),      
    "Klaipėda": (55.703297, 21.144279),
    "Kahramanmaraş": (37.575275, 36.922821),
    "Brașov": (45.657974, 25.601198),
    "Győr": (47.68333, 17.63512)
}

for broken_name, correct_name in name_corrections.items():
    if broken_name in city_coords:
        city_coords.pop(broken_name)     
    city_coords[correct_name] = coordinates[correct_name] 
```

```python
for i,j in city_coords.items():
    print(i,j)
```

### Création d’un DataFrame de coordonnées et fusion avec le DataFrame principal

```python
coords_list = [(city, float(lat), float(lon)) for city, (lat, lon) in city_coords.items()]
coords_df = spark.createDataFrame(coords_list, ["Order City", "New_Latitude", "New_Longitude"])

df = df.join(coords_df, on="Order City", how="left")
```

## Calcul de la distance entre les villes avec la formule de Haversine

```python
from pyspark.sql.functions import col, radians, sin, cos, sqrt, asin
r = 6371

df = df.withColumn("dlat", radians(col("New_Latitude") - col("Latitude"))) \
       .withColumn("dlon", radians(col("New_Longitude") - col("Longitude"))) \
       .withColumn("lat1_rad", radians(col("Latitude"))) \
       .withColumn("lat2_rad", radians(col("New_Latitude"))) \
       .withColumn("a", sin(col("dlat")/2)**2 + cos(col("lat1_rad")) * cos(col("lat2_rad")) * sin(col("dlon")/2)**2) \
       .withColumn("c", 2 * asin(sqrt(col("a")))) \
       .withColumn("distance", col("c") * r) \
       .drop("dlat", "dlon", "lat1_rad", "lat2_rad", "a", "c","New_Latitude","New_Longitude","Longitude","Latitude","Customer City","Order City",)
```
