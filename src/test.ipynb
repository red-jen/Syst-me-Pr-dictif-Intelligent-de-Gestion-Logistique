```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F


spark = SparkSession.builder.appName('test').getOrCreate()


df = spark.read.csv('./data/DataCoSupplyChainDataset.csv', header=True )

df.show(40)

```

```python
df.groupBy('Market').agg(F.avg('Sales'),F.count('Order Id')).collect()
```
